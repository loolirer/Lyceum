# Linear Regression

Linear regression is basically curve fitting but with a fancier name.

The idea here is to model correlated parameters.

---

```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing, model_selection, svm
```

```python
samples = 100

x = np.linspace(0, samples, samples)

A = np.random.randint(-100, 100)
B = np.random.randint(-100, 100)

# Random linear function
y_ground_truth = A * x + B 

# Adding noise to the function
sigma = 500 # Std deviation
y_observed = np.random.normal(y_ground_truth, sigma)
```

```python
# Plotting function
plt.plot(
    x, 
    y_ground_truth,
    label='Ground Truth'
)

plt.scatter(
    x=x, 
    y=y_observed,
    s=4,
    label='Observations'
)

plt.legend(loc='best')
plt.show()
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y_observed, test_size=0.2)
x_train, x_test, y_train, y_test = x_train.reshape(-1, 1), x_test.reshape(-1, 1), y_train.reshape(-1, 1), y_test.reshape(-1, 1)

clf = LinearRegression() # Creating linear regression object
clf.fit(x_train, y_train) # Fitting data to model
accuracy = clf.score(x_test, y_test) # Getting accuracy

print(f'Accuracy: {accuracy*100:.2f}%')

A_estimated = clf.coef_[0][0]
B_estimated = clf.intercept_[0]

y_estimated = A_estimated * x + B_estimated
```

```python
# Plotting function
plt.plot(
    x, 
    y_ground_truth,
    label='Ground Truth'
)

plt.plot(
    x, 
    y_estimated,
    c='red',
    label='Estimated',
)

plt.scatter(
    x=x, 
    y=y_observed,
    s=4,
    label='Observations'
)

plt.legend(loc='best')
plt.show()
```

```python
x_prediction = np.linspace(100, 200, 100).reshape(-1, 1)
y_prediction = np.concatenate(clf.predict(x_prediction))
```

```python
# Plotting function
plt.plot(
    x, 
    y_ground_truth,
    label='Ground Truth'
)

plt.scatter(
    x=x, 
    y=y_observed,
    s=4,
    label='Observations'
)

plt.plot(
    x, 
    y_estimated,
    c='red',
    label='Estimated',
)

plt.plot(
    x_prediction, 
    y_prediction,
    c='green',
    label='Prediction',
)

plt.legend(loc='best')
plt.show()
```
